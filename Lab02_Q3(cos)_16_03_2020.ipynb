```
import numpy as np 
import matplotlib . pyplot as plot

def cos_approx(x, n):
    val = 0
    for m in range(n):
        val += (-1)**m * x**(2 * m) / factorial(2 * m)
        
    return val

def factorial(n):
  if n<2:
        return 1
  else:
        return n*factorial(n-1)

# Get x values of the cos wave
time = np.arange(-3, 3, 0.1);

#Amplitude of the sin wave is cos of a variable like time
amplitude = np.cos(time)

#Amplitude of the cos wave with Tylor Series
amplitude2 = []
for x in time:
  help = 0
  for n in range(2):
    help = help + ((-1)**n)*((x**(2*n))/factorial(2*n))
  amplitude2.append(help)


#Plot a cos wave using time and amplitude obtained for the cos wave
plot.plot(time, amplitude)
plot.plot(time, amplitude2)

#Give a title for the cos wave plot
plot.title('Sine and Taylor approx.')

#Give x axis label for the cos wave plot
plot.xlabel('Time')
#Give y axis label for the cos wave plot
plot.ylabel('Amplitude=cos(time)')
plot.grid(True, which='both')
plot.axhline(y=0, color='k')
plot.show()

#Display the cos wave
plot.show
```

```
#This code shows the difference between amplitudes
amplitude3 = amplitude2 - amplitude
plot.plot(time, amplitude3)
#Give a title for the cos wave plot
plot.title('Difference between numpy cos and Tylor approximation')

#Give x axis label for the cos wave plot
plot.xlabel('Time')
#Give y axis label for the cos wave plot
plot.ylabel('Amplitude')
plot.grid(True, which='both')
plot.axhline(y=0, color='k')
plot.show()
plot.show
```
